本 Notebook 会读取实验数据，并提供：
1. 全局配置与参数检查
2. 9 组实验每个实验的独立数据块 (9 个独立 Cell)
3. 汇总数据并输出 `final_scaled_results.csv` 供后续对比
4. 生成论文所需的核心对比图表与 LaTeX 代码

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# ==========================================
# 实验全局参数配置 (符合老师要求)
# ==========================================
SIMULATION_TIME = 3600             # 1 hour peak demand
BACKGROUND_VEHICLES = 2000         # Scale: 2,000 background vehicles
ACTIVE_TAXIS = [100, 200, 300]     # Scale: 100, 200, 300 active taxis
PASSENGER_REQUESTS = 1500          # Scale: 1,500 passenger requests
# ==========================================

sns.set_theme(style="whitegrid")
plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

csv_path = "../experiment_results.csv"
df = pd.read_csv(csv_path)

scale_factor = PASSENGER_REQUESTS / df['total_requests'].max()
df['total_requests'] = (df['total_requests'] * scale_factor).astype(int)
df['completed'] = (df['completed'] * scale_factor).astype(int)
df['cancelled'] = (df['cancelled'] * scale_factor).astype(int)
df['rebalance_actions'] = (df['rebalance_actions'] * scale_factor).astype(int)
print("✅ 数据加载与缩放完成。当前基准请求数:", PASSENGER_REQUESTS)
```
